```python
import os
import sys

sys.path.append("../..")

import numpy as np
import pandas as pd

from mikasa.io import load_pickle, dump_pickle
```

```python
train = pd.read_csv("../data/raw/train.csv")
test = pd.read_csv("../data/raw/test.csv")
test['Survived'] = np.nan

data = pd.concat([train, test], axis=0)
data["Sex"] = data["Sex"].map({"female": 0, "male": 1})
```

```python
print(data.shape)
data.head()
```

```python
data.groupby('Pclass')[['Age', 'Fare', 'Sex']].agg(['mean', 'std'])
```

```python
data.groupby('Pclass')[['Ticket', 'Cabin']].agg(['nunique'])
```

```python
data.groupby(['Pclass', 'Embarked'])[['Age', 'Fare', 'Sex']].agg(['mean', 'std'])
```

```python
data.groupby(['Pclass', 'Embarked'])[['Ticket', 'Cabin']].agg(['nunique'])
```

```python
data['multi_Sex_Parch'] = (data['Sex'] * data['Parch'])
data['multi_Sex_SibSp'] = (data['Sex'] * data['SibSp'])
data['multi_Sex_Parch+SibSp'] = (data['Sex'] * (data['Parch']+ data['SibSp']))

data['multi_Age_Fare'] = (data['Age'] * data['Fare'])
```
